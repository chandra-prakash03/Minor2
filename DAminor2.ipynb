```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\data_analytics\students_marks_100.csv")
df.head(5)
```

```python
average_marks = df.mean(numeric_only=True)
max_marks = df.max(numeric_only=True)
min_marks = df.min(numeric_only=True)

print("Average Marks:\n", average_marks)
print("\nMaximum Marks:\n", max_marks)
print("\nMinimum Marks:\n", min_marks)
```

```python
average_marks.plot(kind='bar')
plt.title("Average Marks per Subject")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()

```
